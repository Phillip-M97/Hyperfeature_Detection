```python
import pandas as pd
import os
import numpy as np
import json
import matplotlib.pyplot as plt
from PIL import Image
```

```python
csv_annotation = pd.read_csv("annotations/Scaled_Shifted_256.csv")
csv_annotation
```

```python
source_index = 1
geopoints_json = []
source_row = csv_annotation[csv_annotation['Bike index'] == source_index].iloc[0]

source_points = []
for i in range (3, 27, 2):
    source_points.append([int(round(source_row[i])), int(round(source_row[i+1]))])

for index, row in csv_annotation.iterrows():

    target_points = []

    for i in range(3, 27, 2):
        target_points.append([int(round(row[i])), int(round(row[i + 1 ]))])
    
    entry = {
        "source_path": f"{int(source_index)}.png",
        "target_path": f"{int(row['Bike index'])}.png",
        "source_points": source_points,
        "target_points": target_points,
        "category": "unspecified",
        "source_pose": "unspecified",
        "target_pose": "unspecified",
        "mirror": "unspecified",
        "viewpoint_variation": 0,
        "target_bounding_box": "unspecified",
        "source_bounding_box": "unspecified"
    }

    geopoints_json.append(entry)

output_dir = "./annotations"
output_file = "geopoints.json"

full_path = os.path.join(output_dir, output_file)

if not os.path.exists(output_dir):
    os.makedirs(output_dir)

with open(full_path, 'w') as json_file:
    json_file.write('[\n')
    json_file.write(',\n'.join(json.dumps(entry) for entry in geopoints_json))
    json_file.write('\n]')
```

```python
def create_json_entries(csv_annotation, output_dir, output_file):
    geopoints_json = []

    for index, row in csv_annotation.iterrows():
        points = []
        for i in range(3, 27, 2):
            points.append([int(round(row[i])), int(round(row[i + 1]))])

        entry = {
            "image_path": f"{int(row['Bike index'])}.png",
            "image_points": points,
            "category": "unspecified",
            "source_pose": "unspecified",
            "target_pose": "unspecified",
            "mirror": "unspecified",
            "viewpoint_variation": 0,
            "bounding_box": "unspecified"  # Assuming bounding box is the same for simplification
        }

        geopoints_json.append(entry)

    # Save the JSON data
    full_path = os.path.join(output_dir, output_file)
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    
    with open(full_path, 'w') as json_file:
        json_file.write('[\n')
        json_file.write(',\n'.join(json.dumps(entry) for entry in geopoints_json))
        json_file.write('\n]')

# Example usage
create_json_entries(csv_annotation, "./annotations", "geopoints_new.json")
```

```python
with open('annotations/geopoints_short.json', 'r') as file:
    data = json.load(file)

def swap_coords(data):
    for item in data:
        item['image_points'] = [[y,x] for x,y in item['image_points']]
    return data

swapped_data = swap_coords(data)

compact_json = '\n'.join(json.dumps(item, separators=(',', ':')) for item in swapped_data)
output_file_path = 'annotations/geopoints_swapped_2.json'

with open(output_file_path, 'w') as file:
    file.write(compact_json)
```
